```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the data from the Excel file
data = pd.read_excel("Diffraction Grating Data.xlsx")

# Extract the columns for the x, y, and z axes
x = data.iloc[:, 2]
y = data.iloc[:, 1]
z = data.iloc[:, 0]

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)

# Set labels for the axes
ax.set_xlabel('Horizontal (deg)')
ax.set_ylabel('Vertical (deg)')
ax.set_zlabel('Power (mV)')

# Show the plot
plt.show()
```
