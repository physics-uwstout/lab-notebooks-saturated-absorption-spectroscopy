```python
import numpy as np
import math

def create_square_spiral(size, step):
    # Define the coordinates for the square spiral
    x = [0]
    y = [0]
    current_x = 0
    current_y = 0
    direction = 0  # 0: down, 1: left, 2: up, 3: right
    expansion = 0

    for i in range(size):
        if direction == 0:
            for k in range(math.floor(expansion)):
                current_y -= 1
                x.append(current_x)
                y.append(current_y)
            expansion += 0.5
        elif direction == 1:
            for k in range(math.floor(expansion)):
                current_x -= 1 * math.floor(expansion)
                x.append(current_x)
                y.append(current_y)
            expansion += 0.5
        elif direction == 2:
            for k in range(math.floor(expansion)):
                current_y += 1 * math.floor(expansion)
                x.append(current_x)
                y.append(current_y)
            expansion += 0.5
        elif direction == 3:
            for k in range(math.floor(expansion)):
                current_x += 1 * math.floor(expansion)
                x.append(current_x)
                y.append(current_y)
            expansion += 0.5

        direction = (direction + 1) % 4

    return np.array([x, y])

# Example usage
spiral = create_square_spiral(3, 1)
print(spiral)

plt.plot(spiral[0], spiral[1])
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

```

```python
# Define the coordinates for the square spiral
x = [0]
y = [0]
current_x = 0
current_y = 0
direction = 0  # 0: down, 1: left, 2: up, 3: right
expansion = 0
step = 1
size = 16*step

while (y[-1] != size and x[-1] != size):
    if direction == 0:
        for k in range(math.floor(expansion)):
            current_y += step
            x.append(current_x)
            y.append(current_y)
        expansion += 0.5
    elif direction == 1:
        for k in range(math.floor(expansion)):
            current_x += step
            x.append(current_x)
            y.append(current_y)
        expansion += 0.5
    elif direction == 2:
        for k in range(math.floor(expansion)):
            current_y -= step
            x.append(current_x)
            y.append(current_y)
        expansion += 0.5
    elif direction == 3:
        for k in range(math.floor(expansion)):
            current_x -= step
            x.append(current_x)
            y.append(current_y)
        expansion += 0.5

    direction = (direction + 1) % 4

spiral = np.array([x, y])

plt.plot(spiral[0], spiral[1])
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

df_spiral = pd.DataFrame({'X': spiral[0], 'Y': spiral[1]})
df_spiral.to_excel('Diffraction Data Test 2.xlsx', index=False)

```

```python
import math
#Temperature Calculation

# Constants for the Steinhart-Hart equation
a = 3.3540170E-03 
b = 2.5617244E-04
c = 2.1400943E-06
d =-7.2405219E-08

Ar = -1.5470381E+01
Br = 5.6022839E+03 
Cr = -3.7886070E+05 
Dr = 2.4971623E+07

# Resistance of the thermistor at 25°C
R_25 = 10e3

# Temperature calculation function
def calculate_temperature(resistance):
    # Calculate the reciprocal of the resistance
    reciprocal_resistance = resistance / R_25

    temperature = 1 / (a + b * math.log(reciprocal_resistance) + c * math.pow(math.log(reciprocal_resistance), 2) + d * math.pow(math.log(reciprocal_resistance), 3)) - 273.15

    return temperature

def calculate_resistance(temperature):
    # Calculate the reciprocal of the resistance
    
    resistance = R_25 * math.exp(Ar + Br/temperature + Cr / math.pow(temperature, 2) + Dr / math.pow(temperature, 3))

    return resistance

print("Set resistance for temp:",calculate_resistance(15+273.15))

print("Actual Temp:",calculate_temperature(12.1e3))
```
